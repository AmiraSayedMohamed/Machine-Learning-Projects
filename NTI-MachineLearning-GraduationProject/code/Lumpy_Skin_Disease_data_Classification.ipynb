## Remove outliers

```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Removing the outliers
for col in numeric_df:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + (1.5 * iqr)
    lower_limit = q1 - (1.5 * iqr)
    new_df2 = df.loc[(df[col] <= upper_limit) & (df[col] >= lower_limit)]
print('before removing outliers: ', len(df))
print('after removing outliers: ', len(new_df2))
print('outliers:', len(df)-len(new_df2))
```

Interpretable model.

```python
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, roc_auc_score # ROC Curve (Receiver Operating Characteristic Curve)
#  ROC curve is a plot of the True Positive Rate (TPR) versus the False Positive Rate
from sklearn.preprocessing import StandardScaler

# Oversample minority class using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Logistic Regression model with class weights to handle imbalance
lo = LogisticRegression(class_weight='balanced', random_state=42)
lo.fit(X_train, y_train)

y_pred_lo = lo.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_lo)
print(f"Accuracy: {accuracy}")
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_lo))
```

# Support Vector Machines (SVM)

- Best for: High-dimensional datasets where you want to find a decision boundary that maximizes the margin between classes.

- How it works: SVM tries to find the hyperplane that best separates the two classes. It works well in high-dimensional spaces and when the decision boundary is not necessarily linear.

- Pros:
Effective in high-dimensional spaces.
Works well with both linear and non-linear boundaries.

- Cons:
Can be slow to train, especially with large datasets.
The choice of the kernel (linear, radial basis function, etc.) can impact performance.

```python
from sklearn.svm import SVC

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# SVM Model
clf_svm = SVC(random_state=42, kernel='linear', probability=True)
clf_svm.fit(X_train, y_train)

# Predict and Evaluate
y_pred_svm = clf_svm.predict(X_test)
y_pred_proba_svm = clf_svm.predict_proba(X_test)[:, 1]

# Model Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba_svm))
```

# K-Nearest Neighbors (KNN)

- Best for: Simpler problems where you want to make predictions based on the closest data points.

- How it works: KNN classifies a data point based on the majority class of its nearest neighbors in the feature space. It doesn't learn a model explicitly but instead memorizes the training data.

- Pros:
Simple to implement.
Non-parametric, meaning no assumption about the data distribution.

- Cons:
Can be slow for large datasets since it needs to compute distances between all points during inference.
Sensitive to irrelevant features and the choice of distance metric.

```python
from sklearn.neighbors import KNeighborsClassifier

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# KNN Model
clf_knn = KNeighborsClassifier(n_neighbors=5)
clf_knn.fit(X_train, y_train)

# Predict and Evaluate
y_pred_knn = clf_knn.predict(X_test)
y_pred_proba_knn = clf_knn.predict_proba(X_test)[:, 1]

# Model Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba_knn))
```

# Neural Networks (Deep Learning)

- Best for: Complex relationships between features that might be difficult for traditional models to capture.

- How it works: Neural networks are composed of layers of nodes (neurons) that process and transform input data. They are particularly useful for complex, high-dimensional data like images, text, and even structured data like yours.

- Pros:
Can capture complex, non-linear relationships between features.
Suitable for large datasets.

- Cons:
Require large amounts of data to perform well.
Computationally expensive and may take longer to train.
Less interpretable than tree-based models.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Neural Network Model
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Predict and Evaluate
y_pred_nn = (model.predict(X_test) > 0.5).astype(int)
y_pred_proba_nn = model.predict(X_test)

# Model Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred_nn))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba_nn))
```

# Naive Bayes
- Best for: When the features are conditionally independent and for simpler problems.

- How it works: Naive Bayes is a probabilistic classifier that assumes each feature is independent of the others given the class. It calculates the probability of each class given the input features and assigns the class with the highest probability.

- Pros:
Fast and simple to implement.
Works well with high-dimensional data.

- Cons:
The independence assumption may not hold true, leading to suboptimal performance if features are correlated.

```python
from sklearn.naive_bayes import GaussianNB
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Naive Bayes Model
clf_nb = GaussianNB()
clf_nb.fit(X_train, y_train)

# Predict and Evaluate
y_pred_nb = clf_nb.predict(X_test)
y_pred_proba_nb = clf_nb.predict_proba(X_test)[:, 1]

# Model Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred_nb))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba_nb))
```

# Decision Trees

- Best for: Interpretable models where you want to understand how decisions are being made.

- How it works: A decision tree splits the data into subsets based on the value of a feature, recursively partitioning the data until each subset is pure enough to make a prediction.

- Pros:
Easy to interpret.
Can handle both numerical and categorical data.

- Cons:
Prone to overfitting, especially with deep trees.

```python
from sklearn.tree import DecisionTreeClassifier

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Decision Tree Model
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt.fit(X_train, y_train)

# Predict and Evaluate
y_pred_dt = clf_dt.predict(X_test)
y_pred_proba_dt = clf_dt.predict_proba(X_test)[:, 1]

# Model Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba_dt))
```
